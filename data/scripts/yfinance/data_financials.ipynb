```python
# Pull into memory
data_dictionary = dict()
for stock in seed_ind_stock_list:
    try:
        data_dictionary[stock] = pd.read_csv(f"./temp/{stock}_financials.csv")
    except Exception as e:
        print(f"Failed to read {stock}:", e)
```

```python
# Analyze indices
common_indices = None
for k, v in data_dictionary.items():
    row_names = set(v["Unnamed: 0"])
        
    if "Basic EPS" not in row_names:
        print(k, "has no Basic EPS")
        continue

    if common_indices == None:
        common_indices = row_names
    else:
        common_indices = common_indices.intersection(row_names)
```

```python
common_indices
```

```python
# Pivot each DF
df_test = data_dictionary["A"][data_dictionary["A"]["Unnamed: 0"].isin(["Basic EPS", "Diluted EPS", "Total Revenue", "Net Income"])]
```

```python
melted = df_test.melt(id_vars=["Unnamed: 0"])
melted["Symbol"] = "v"
melted
```

```python
# Collect columns: Basic EPS, Total Revenue, Net Income
df_financials = None
is_set = False
for k, v in data_dictionary.items():
    df_subset = v[v["Unnamed: 0"].isin(["Basic EPS", "Diluted EPS", "Total Revenue", "Net Income"])].copy()
    df_subset = df_subset.melt(id_vars=["Unnamed: 0"])
    df_subset["Symbol"] = k
    
    if not is_set:
        df_financials = df_subset.copy()
        is_set = True
    else:
        df_financials = pd.concat([df_financials, df_subset])
        
df_financials.head(20)
```

```python
# Rename columns
df_financials.rename(columns={
    "Unnamed: 0": "item",
    "variable": "quarter_ending_on",
    "Symbol": "symbol"
}, inplace=True)

# Reorder
df_financials_table = df_financials[["symbol", "quarter_ending_on", "item", "value"]]
df_financials_table
```
