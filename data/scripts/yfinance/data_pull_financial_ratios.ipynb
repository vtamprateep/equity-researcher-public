```python
# Import
import yfinance as yf
import pandas as pd
```

```python
# Import seed data
df_seed = pd.read_csv("../sector_etf_holding.csv")
```

```python
df_seed.head()
```

```python
data = yf.download("SPY AAPL", period="1mo")
```

```python
# Merge holdings into space separate string for download
holdings = set(df_seed["holding"])
holdings.update(set(df_seed["etf_name"]))

space_del_holdings = " ".join(list(holdings))
space_del_holdings
```

```python
data_holdings = yf.download(space_del_holdings, period="1y")
```

```python
df_hold_stack = data_holdings.stack(level=1)
df_ungrouped_stack = df_hold_stack.reset_index()
df_ungrouped_stack
```

```python
column_name = {
    "Date": "close_date",
    "level_1": "symbol",
    "Adj Close": "adj_close",
    "Close": "close",
    "High": "high",
    "Low": "low",
    "Open": "open",
    "Volume": "volume"
}
df_ungrouped_stack.rename(columns=column_name, inplace=True)
df_ungrouped_stack
```

```python
# Charts table
df_charts_table = df_ungrouped_stack.copy()
```

```python
# Make symbols table
df_symbol_table = df_ungrouped_stack["symbol"].drop_duplicates()
df_symbol_table = df_symbol_table.reset_index()
df_symbol_table.index += 1
df_symbol_table = df_symbol_table.reset_index()
df_symbol_table
```

```python
df_symbol_table.drop(columns=["index"], inplace=True)
df_symbol_table.rename(columns={
    "level_0": "id"
}, inplace=True)
```

```python
df_symbol_table

# Save
df_symbol_table.to_csv("db_symbol_seed.csv", index=False)
```

```python
df_symbol_table
```

```python
# Tie relation between charts and symbol
df_merged = pd.merge(df_symbol_table, df_charts_table, on="symbol", how="left")
df_merged
```

```python
# Drop symbol, rename
df_charts_table = df_merged.drop(columns=["symbol"])
df_charts_table.rename(columns={"id": "symbol_id"}, inplace=True)

# Add ID column
df_charts_table.index += 1
df_charts_table.reset_index(inplace=True)
df_charts_table.rename(columns={"index": "id"}, inplace=True)

# Re-order
df_charts_table = df_charts_table[["id", "symbol_id", "close_date", "open", "high", "low", "close", "adj_close"]]
df_charts_table
```

```python
df_charts_table.to_csv("db_charts_seed.csv", index=False)
```

# Create Holdings Table

```python
df_initial_holdings = pd.read_csv("../sector_etf_holding.csv")
df_symbol = pd.read_csv("../../seed/db_symbol_seed.csv")
df_charts = pd.read_csv("../../seed/db_symbol_seed.csv")
```

```python
# Replace holding fund with IDs
df_initial_holdings.rename(columns={"etf_name": "symbol"}, inplace=True)
df_holding = pd.merge(df_initial_holdings, df_symbol, on="symbol", how="left")
df_holding.drop(columns=["symbol"], inplace=True)
df_holding.rename(columns={"id": "holding_symbol_id"}, inplace=True)

df_holding.head()
```

```python
# Replace held symbol ID
df_holding.rename(columns={"holding": "symbol"}, inplace=True)
df_holding = pd.merge(df_holding, df_symbol, on="symbol", how="left")
df_holding.drop(columns=["symbol"], inplace=True)
df_holding.rename(columns={"id": "symbol_id"}, inplace=True)

df_holding.head()
```

```python
df_holding.dropna(inplace=True)
```

```python
df_holding.to_csv("../../seed/db_holdings_seed.csv", index=False)
```
